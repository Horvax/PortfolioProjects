```python
df_tracks = pd.read_csv ('D:/work/Spotify Project/tracks.csv')
df_tracks.set_index("release_date", inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
# Converting the duration from miliseconds to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis =1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only = True)
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", 
            scatter_kws = {"color":"c"}, line_kws = {"color":"green"}).set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", 
            scatter_kws = {"color":"green"}, line_kws = {"color":"magenta"}).set(title = "Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Number of songs per year")
```

```python
# looking at the dependancy of the song duration and a year

total_duration = df_tracks.duration
fig_dimensions = (18,7)
fig, ax = plt.subplots(figsize = fig_dimensions)
fig = sns.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```

```python
sns.set_style(style="whitegrid")
fig_dimensions = (10,5)
fig, ax = plt.subplots(figsize = fig_dimensions)
fig = sns.lineplot(x = years, y = total_duration, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 60)
```

```python
# Opening a genre file

df_genre = pd.read_csv("D:/work/Spotify Project/SpotifyFeatures.csv")
```

```python
df_genre.head()
```

```python
# looking at the duration of the songs in different genres

plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
# We have found out that we have 2 children's music genres, so we need to clean that up

df_genre.genre.unique()
```

```python
# Replacing the genre's names

df_genre = df_genre.replace(['Children’s Music'], "Children's Music")
```

```python
# Checking if the replacement has been made properly

df_genre.genre.unique()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = "Top 5 genres by popularity")
```
